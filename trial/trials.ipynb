```python
from src.sudoku import Sudoku
import csv
import os
```

```python
simple_path = '/home/jgeng/Documents/Git/ECE637-Project1/input/simple.csv'
easy_path = '/home/jgeng/Documents/Git/ECE637-Project1/input/easy.csv'
medium_path = '/home/jgeng/Documents/Git/ECE637-Project1/input/medium.csv'
hard_path = '/home/jgeng/Documents/Git/ECE637-Project1/input/hard.csv'
sukudo_data_paths = [simple_path, easy_path, medium_path, hard_path]
modes = ['bfs']
```

```python
def write_results(fp, meta_data):
    
    header = ['givens', 'singles', 'hidden_singles', 'naked_pairs', 'hidden_paris', 'triples', 'intersections', 
              'guesses','backtracks', 'difficulty', 'time', 'max_space']
    with open(fp, mode='w') as csv_file:
        writer = csv.DictWriter(csv_file, fieldnames=header, extrasaction='ignore')
        writer.writerows(meta_data)
        
def read_results(fp):
    
    out = []
    header = ['givens', 'singles', 'hidden_singles', 'naked_pairs', 'hidden_paris', 'triples', 'intersections', 
              'guesses','backtracks', 'difficulty', 'time', 'max_space']
    with open(fp, mode='r') as csv_file:
        reader = csv.DictReader(csv_file, fieldnames=header)
        for row in reader:
            out.append(row)
    return out
```

```python
solver = Sudoku()
large_path = '/home/jgeng/Documents/Git/ECE637-Project1/input/large.csv'
parsed = Sudoku.parser(large_path)
large = parsed[1]
solver.read_sudoku(large)
print(solver.solve_sudoku(mode='bfs', check_correct=False))
```

```python
for trial_data_path in sukudo_data_paths:
    difficulty = trial_data_path[trial_data_path.find('input/') + 6:-4]
    for mode in modes:
        out_path = '/home/jgeng/Documents/Git/ECE637-Project1/output/{}_{}.csv'.format(difficulty, mode)
        parsed_data = Sudoku.parser(trial_data_path)
        results = []
        max_space = 351 if mode != 'bfs' else 0
        for sudoku_meta_data in parsed_data:
            solver.read_sudoku(sudoku_meta_data)
            has_solution, time = solver.solve_sudoku(mode=mode, check_correct=False)
            result = solver.meta_data
            result['time'] = time
            result['max_space'] = max_space if mode != 'bfs' else solver.max_qsize
            results.append(result)
            print("Solved a {} suduku with {} approach in {} using {} queue items".format(difficulty, mode, time, solver
                                                                                .max_qsize))
        print("Trial {}_{} completed!".format(difficulty, mode))
        write_results(fp=out_path, meta_data=results)
            
            
```
